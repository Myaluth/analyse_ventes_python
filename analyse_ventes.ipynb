```python
import pandas as pd

data = {
    "date": ["2026-01-05", "2026-01-12", "2026-02-03", "2026-02-10", "2026-03-01", "2026-03-15"],
    "produit": ["Laptop", "Souris", "Laptop", "Clavier", "Écran", "Souris"],
    "catégorie": ["Informatique", "Accessoires", "Informatique", "Accessoires", "Informatique", "Accessoires"],
    "quantité": [5, 10, 3, 10, 8, 8],
    "prix_unitaire": [1200, 25, 1200, 45, 300, 25],
}

df = pd.DataFrame(data)
df
```

```python
df["chiffre_affaires"] = df["quantité"] * df["prix_unitaire"]
df
```

```python
ca_total = df["chiffre_affaires"].sum()
ca_total
ca_par_produit = df.groupby("produit")["chiffre_affaires"].sum().sort_values(ascending=False)
ca_par_produit
```

```python
import matplotlib.pyplot as plt

ca_par_produit.plot(kind="bar", title="Chiffre d'affaires par produit")
plt.ylabel("Chiffre d'affaires (€)")
plt.show()
```
